```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use('seaborn')
```

```python
gapminder = pd.read_csv('gapminder_data.csv')
gapminder.continent=gapminder.continent.astype('category')
```

```python
gapminder.head()
```

```python
gapminder
```

```python
#ax = plt.gca()
colors={x:i for i,x in enumerate(gapminder['continent'].unique())}
gapminder['color']=gapminder['continent'].map(colors).astype('int')
gapminder.plot(kind='scatter',c='color',s=2,logx=True,colormap='viridis',y='lifeExp',colorbar=False,x='gdpPercap',legend=True)
```

```python
gapminder['loggdpPerCap']=gapminder['gdpPercap'].map(np.log10)
ax=sns.lmplot(x='loggdpPerCap',y='lifeExp',hue='continent',scatter_kws={'s':10},data=gapminder)
```
